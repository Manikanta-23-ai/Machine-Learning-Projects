IMPORTING THE ALL NECCESSARY LIBRIARIES

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

READING THE DATA

```python
df =pd.read_csv("/content/df_test.csv")
df.head()
```

```python
df.tail()
```

This section prepares our data for machine learning by separating input features from the target, removing incompatible data types, and identifying which columns need special mathematical treatment.

Input Features** - All information we use to MAKE PREDICTIONS (everything except 'grade')

Target Variable** - What we want to PREDICT (in this case, 'grade')
3. **Numeric Features** - Columns with numbers (bedrooms=2, price=305000) that can be directly standardized
4. **Categorical Features** - Columns with True/False values that need conversion to 0/1
5. **StandardScaler** - Makes all numbers fit on similar scale (prevents large numbers from dominating)
6. **OneHotEncoder** - Converts True/False to 1/0 for model compatibility

```python
X = df.drop("grade", axis=1)
X = X.drop("date", axis=1)
y = df["grade"]
num_features = [col for col in X.columns if X[col].dtype in ['int64', 'float64']]
cat_features = [col for col in X.columns if X[col].dtype == 'bool']

preprocessor = ColumnTransformer([
    ("num", StandardScaler(), num_features),
    ("cat", OneHotEncoder(drop="first"), cat_features)
])
```

USE PIPELINES FOR OPERATIONS

```python
model_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("regressor", LinearRegression())
])
```

TRAIN AND TEST DATA BY USING TRAIN AND TEST SPILT FOR LINEAR REGRESSION

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=43)
```

CALCULATING THE ABSOLUTE ERROR AND R SQUARE

```python
model = LinearRegression()
```

```python
model_pipeline.fit(X_train, y_train)
predictions = model_pipeline.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print(f'Mean Absolute Error: {mae}')
print(f'R² Score: {r2}')
```
